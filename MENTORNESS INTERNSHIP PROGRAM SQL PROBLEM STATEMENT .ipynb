# 1. TOTAL NUMBER OF RESERVATION IN THE DATASET

```python
%%sql
SELECT
    COUNT(Booking_ID) as total_number_of_reservations
FROM
    mentorness_internship_project.`hotel reservation dataset`;
```

# 2. MOST POPULAR MEAL PLAN AMONG GUESTS

```python
%%sql
SELECT type_of_meal_plan, COUNT(type_of_meal_plan) AS meal_plan_count
FROM mentorness_internship_project.`hotel reservation dataset`
GROUP BY type_of_meal_plan
ORDER BY meal_plan_count DESC
LIMIT 1;
```

# 3. AVERAGE PRICE PER ROOM FOR RESERVATIONS INVOLVING CHILDREN

```python
%%sql
SELECT 
    AVG(avg_price_per_room)
FROM 
    mentorness_internship_project.`hotel reservation dataset`
WHERE no_of_children >0
```

# 4. THE NUMBER OF RESERVATIONS MADE IN 2018

```python
%%sql
SELECT
    COUNT(Booking_ID) as Number_of_reservations_in_2018
FROM 
    mentorness_internship_project.`hotel reservation dataset`
WHERE arrival_date LIKE '%2018'
```

# 5. MOST COMMONLY BOOKED ROOM TYPE

```python
%%sql
SELECT room_type_reserved, COUNT(room_type_reserved) as room_count
FROM mentorness_internship_project.`hotel reservation dataset`
GROUP BY room_type_reserved
ORDER BY room_count DESC
LIMIT 1;
```

# 6. NUMBER OF RESERVATIONS THAT FALL ON A WEEKEND

```python
%%sql
SELECT
    COUNT(Booking_ID) as Reservations_on_weekend
FROM mentorness_internship_project.`hotel reservation dataset`
WHERE no_of_weekend_nights >0
```

# 7. HIGHEST AND LOWEST LEAD TIME FOR RESERVATIONS

```python
%%sql
SELECT MAX(lead_time) as Highest_lead_time, MIN(lead_time) as lowest_lead_time
FROM mentorness_internship_project.`hotel reservation dataset`
```

# 8. MOST COMMON MARKET SEGMENT TYPE FOR RESERVATIONS

```python
%%sql
SELECT
    market_segment_type, COUNT(market_segment_type) as market_segment_type_count
FROM mentorness_internship_project.`hotel reservation dataset`
GROUP BY market_segment_type
ORDER BY market_segment_type_count DESC
LIMIT 1;
```

# 9. NUMBER OF RESERVATIONS WITH BOOKING STATUS AS 'CONFIRMED'

```python
%%sql
SELECT
    COUNT(Booking_ID) as Confirmed_Reservations
FROM mentorness_internship_project.`hotel reservation dataset`
WHERE booking_status LIKE 'Confirmed'
```

# 10. TOTAL NUMBER OF ADULTS AND CHILDREN ACROSS ALL RESERVATIONS

```python
%%sql
SELECT
    SUM(no_of_adults) as Number_of_adults, SUM(no_of_children) as Number_of_children
FROM mentorness_internship_project.`hotel reservation dataset`
```

# 11. AVERAGE NUMBER OF WEEKEND NIGHTS FOR RESERVATIONS INVOLVING CHILDREN

```python
%%sql
SELECT
    AVG(no_of_weekend_nights) as Weekend_nights_involving_children
FROM mentorness_internship_project.`hotel reservation dataset`
WHERE no_of_children >0
```

# 12. RESERVATIONS MADE IN EACH MONTH OF THE YEAR

```python
%%sql
SELECT 
    MonthName,
    COUNT(*) AS Reservations
FROM (
    SELECT 
        CASE SUBSTRING(arrival_date, 4, 2)
            WHEN '01' THEN 'January'
            WHEN '02' THEN 'February'
            WHEN '03' THEN 'March'
            WHEN '04' THEN 'April'
            WHEN '05' THEN 'May'
            WHEN '06' THEN 'June'
            WHEN '07' THEN 'July'
            WHEN '08' THEN 'August'
            WHEN '09' THEN 'September'
            WHEN '10' THEN 'October'
            WHEN '11' THEN 'November'
            WHEN '12' THEN 'December'
        END AS MonthName
    FROM 
        mentorness_internship_project.`hotel reservation dataset`
) AS subquery
GROUP BY 
    MonthName
ORDER BY 
    FIELD(MonthName, 'January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December');
```

# 13. AVERAGE NUMBER OF NIGHTS SPENT BY GUESTS FOR EACH ROOM TYPE

```python
%%sql
SELECT 
    room_type_reserved,
    AVG(no_of_weekend_nights + no_of_week_nights) AS Avg_Nights
FROM 
    mentorness_internship_project.`hotel reservation dataset`
GROUP BY 
    room_type_reserved;
```

# 14. COMMON ROOM TYPE AND AVERAGE PRICE FOR THE ROOM TYPE FOR THE RESERVATIONS INVOLVING CHILDREN

```python
%%sql
SELECT 
    room_type_reserved,
    AVG(avg_price_per_room) AS Avg_Price
FROM 
    mentorness_internship_project.`hotel reservation dataset`
WHERE 
    no_of_children > 0
GROUP BY 
    room_type_reserved
ORDER BY 
    COUNT(*) DESC
LIMIT 1;
```

# 15. MARKET SEGMENT TYPE THAT GENERATES THE HIGHEST AVERAGE PRICE PER ROOM

```python
%%sql
SELECT 
    market_segment_type,
    AVG(avg_price_per_room) AS Avg_Price_Per_Room
FROM 
    mentorness_internship_project.`hotel reservation dataset`
GROUP BY 
    market_segment_type
ORDER BY 
    Avg_Price_Per_Room DESC
LIMIT 1;
```
